```python
##Assignment 1 Stuff
#Load Packages
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
#Generate Data
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0, 100,50)

#Plot Data
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y)
```

```python
#Tweak Axis
fig, ax = plt.subplots(figsize=(5, 3))
ax.barh(x,y) #bar is "normal", barh flips the axes
```

```python
#Make Basic Plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
```

```python
#Turn into histogram
fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(y)
```

```python
#Adding titles
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)

ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
fig.tight_layout()
```

```python
#Define Fonts and add them
font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_title('Total growth over time', fontdict = font1)
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
fig.tight_layout()
```

```python
#Shift title left
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_title('Total growth over time', fontdict = font1, loc = 'left') #shifts title to left
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
fig.tight_layout()
```

```python
#Modifying data points to be stars
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter( x,
    y,
    marker='*', # NEW ADDITIONS
    color = "indigo") # NEW ADDITIONS
```

```python
#Adding dashed lines
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,
    y,
    marker='*',
    color = 'indigo', # Can also use hex (e.g., #7425b9)
    linestyle = '--', # NEW ADDITIONS
    linewidth = 2) # NEW ADDITIONS
```

```python
#Tweaking point colouring
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y,marker='*',
    markersize = 12, # NEW ADDITIONS
    color = '#7425b9',
    linestyle = '--',
    linewidth = 2,
    markeredgecolor = '#fa9359', # NEW ADDITIONS; outline colours
    markerfacecolor = '#000000' ) # NEW ADDITIONS; inner colours
ax.grid(color = 'blue',
        linestyle = '-.') #Adds grid lines and can tweak them
```

```python
#Trying To replicate graph
import matplotlib.pyplot as plt
import seaborn as sns
#Load data
df = sns.load_dataset('iris')
```

```python
#Creating topography graph
sns.set_style("white")
sns.kdeplot(x=df.sepal_width, y=df.sepal_length)
plt.show()
```

```python
#Color changes to topography
sns.set_style("white")
sns.kdeplot(x=df.sepal_width, y=df.sepal_length, cmap="Reds", fill=True)
plt.show()
```

```python
#Smoothing graph
sns.set_style("white")
sns.kdeplot(x=df.sepal_width, y=df.sepal_length,
            cmap="Blues", fill=True, bw_adjust=0.5)
plt.show()

#Does visualization succeed accross the "Three" areas?
    #Aesthetic: The final product is okay, but a little unappealing to me - these graph styles are foreign to me
        #which is why I chose it. But, the colouring and smoothing is okay - I prefer it less smooth.
    #Substantial: It's accurate/honest regarding the classic dataset
    #Perceptual: The information is somewhat clear, but I would still prefer a "basic" scatterplot with a line of best fit.
```
